## Conditional Filtering

```python
import numpy as np
import pandas as pd


df = pd.read_csv('tips.csv')
df.head()
```

### Conditions

```python
df['total_bill'] > 30
```

```python
# Get the data where the total_bill is higher than 30
bool_series = df['total_bill'] > 30
df[bool_series].head()
```

```python
df[df['total_bill'] > 30].head()
```

```python
df[df['sex'] == 'Male'].head()
```

### Multiple Conditions

Steps:
- Get the conditions
- Wrap each condition in parenthesis
- Use the `|` or `&` operator, depending if you want an 
    - `OR` | (either condition is True)
    - `AND` & (both conditions must be True)
- You can also use the `~` operator as a `NOT` operation

```python
df[(df['total_bill'] > 30) & (df['sex'] == 'Male')].head()
```

```python
df[(df['total_bill'] > 30) & ~(df['sex'] == 'Male')].head()
```

```python
df[(df['total_bill'] > 30) & (df['sex'] != 'Male')].head()
```

```python
# Data from the Weekend
df[(df['day'] == 'Sat') | (df['day'] == 'Sun')].head()
```

`isin()`operator to filter by a list of options.

```python
days = ['Sat', 'Sun']
df['day'].isin(days)
```

```python
df[df['day'].isin(days)].head()
```
